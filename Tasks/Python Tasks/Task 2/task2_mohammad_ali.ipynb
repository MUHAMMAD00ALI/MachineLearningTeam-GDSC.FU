```python

lower_number = 0
upper_number = 90


print("Prime numbers between", lower_number, "and", upper_number, "are:")

for num in range(lower_number, upper_number + 1):
   # all prime numbers are greater than 1
   if num > 1:
    for i in range(2, num):
           if (num % i) == 0:
                break
else:
           print(num)
        
            
```
